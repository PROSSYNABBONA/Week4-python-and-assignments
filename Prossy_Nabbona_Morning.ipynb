```python
#pandas Correlation
#function called corr()
import pandas as pd
#create a dataframe
df={
    'my_values1':(1,2,3,4,5),
    'my_values2':(5,4,3,1,1),
    'my_values3':(6,7,8,9,10)
    }
df=pd.DataFrame(df)
#compute cororrelation
correlation=df.corr()
#printing output
print(correlation)
# +1 and -1 it is a perfect correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# create a correlation heatmap
sns.heatmap(correlation,annot=True,cmap='coolwarm')
# display the heatmap
plt.show
```
